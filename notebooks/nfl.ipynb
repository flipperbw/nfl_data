```python
base_url = 'http://www.foxsports.com'
add_url = '/nfl/transactions?year={}&month={}&type={}'
year = 2017
month_range = range(8,9)
typ = 0

pattern = re.compile(r'Next')
```

```python
def print_data(url):
    q = requests.get(url).text
    soup = bs(q, "lxml")
    
    r = soup.select('section.wisbb_body')
    if not len(r):
        print 'ERROR: could not find body'
        return False
    else:
        r = r[0]
    
    trans = r.select('#wisbb_transactions')
    if not len(trans):
        print 'ERROR: could not find transactions'
        return False
    else:
        trans = trans[0]
    
    trs = trans.find('tbody').find_all('tr')
    for tr in trs:
        tds = tr.find_all('td')
        team_spans = tds[0].find('span').find_all('span')
        team_name = team_spans[0].text.strip()
        team_abbrev = team_spans[1].text.strip()

        name_spans = tds[1].find('span').find_all('span')
        player_name = name_spans[0].text.strip()
        player_pos = name_spans[2].text.strip()

        trans_type = tds[2].text.strip()

        trans_date = tds[3].text.strip()

        print '{}|{}|{}|{}|{}|{}'.format(team_name, team_abbrev, player_name, player_pos, trans_type, trans_date)
    
    pags = r.select('.wisbb_paginator')
    if not len(pags):
        return False
    else:
        pags = pags[0]
        next_page = pags.find('a', text=pattern)
        if next_page:
            href = next_page['href']
            if href:
                return print_data(base_url + href)

    return
    
```

```python
for month in month_range:
    url = base_url + add_url.format(year, month, typ)
    print_data(url)
```

```python
p = requests.get('https://www.fantasypros.com/nfl/depth-charts.php').text
sp = bs(p)
```

```python
main = sp.select('.main-content .inner')[0]
ch = main.find_all('div', {'class': 'team-list'}, recursive=False)
for c in ch:
    team = c.find('input')['value']
    
    pos = c.select('.position-list')
    for p in pos:
        po = p.select('.position-head')[0].text.strip().replace('ECR','')
        
        players = p.find('ul').find_all('li')
        for pl in players:
            pl_divs = pl.find_all('div')
            
            rnk = pl_divs[0].text
            name = pl_divs[1].text

            print '{}|{}|{}|{}'.format(team, po, rnk, name)
```
